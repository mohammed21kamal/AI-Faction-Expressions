```python
import pandas as pd
import numpy as np
```

```python
ratings = pd.read_csv("ratings.csv")
movies = pd.read_csv("movies.csv")
```

```python
movie_ratings = ratings.merge(movies, on='movieId')
movie_ratings['title'].value_counts()[0:10]
```

```python
movie_ratings.head(20)
```

```python
movie_ratings['genres'] = movie_ratings['genres'].apply(lambda x: x.split("|"))
movie_ratings.head()
```

```python
movie_ratings = movie_ratings[['userId', 'movieId', 'genres']]
movie_ratings = movie_ratings.sort_values(by="userId", ascending=True)
movie_ratings.head(100)
```

```python
from collections import Counter

genre_frequency = Counter(g for genres in movie_ratings['genres'] for g in genres)

print(f"There are {len(genre_frequency)} genres.")
genre_frequency

```

```python
num_unique_users = movie_ratings['userId'].nunique()

print(f"There are {num_unique_users} unique users.")
```

```python
user_of_count_rates_movies = ratings.groupby('userId')['rating'].agg(['count'])
user_of_count_rates_movies.head()
```

```python
import pandas as pd

# Ù‚Ø±Ø§Ø¡Ø© Ù…Ù„Ù Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª
movie_ratings = pd.read_csv('movie_ratings.csv')

# Ø§Ù„Ø£Ù†ÙˆØ§Ø¹ Ø§Ù„Ù…Ù…ÙƒÙ†Ø©
genres = sorted(set(g for G in movie_ratings['genres'] for g in G))

# Ø¥Ø¶Ø§ÙØ© Ø¹Ù…ÙˆØ¯ Ù„ÙƒÙ„ Ù†ÙˆØ¹
for g in genres:
    movie_ratings[g] = movie_ratings.genres.transform(lambda x: int(g in x))

# Ø¥Ø²Ø§Ù„Ø© Ø§Ù„Ø£Ø¹Ù…Ø¯Ø© ØºÙŠØ± Ø§Ù„Ø¶Ø±ÙˆØ±ÙŠØ©
movie_genres = movie_ratings.drop(columns=['movieId', 'genres'])

# Ø·Ø¨Ø§Ø¹Ø© Ø§Ù„Ù†ØªØ§Ø¦Ø¬ Ù…Ø±ØªØ¨Ø© Ø£Ø¨Ø¬Ø¯ÙŠÙ‹Ø§
sorted_movie_genres = movie_genres.sort_values(by=genres)
print(sorted_movie_genres)
```
